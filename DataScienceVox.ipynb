```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
# Downloading vox.com
headers = {'Accept-Languange': 'en-US,en;q=0.8'}
url = 'https://www.vox.com/'
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Extract the number of titles
newstitle = soup.select('h2.c-entry-box--compact__title')
print(newstitle[0])
```

```python
# Get Links
text = soup.select('h2 a')
print(text)
```

```python
# Make link clean
links=[]
for a in soup.select('h2 a'):
  links.append(a.attrs.get('href'))

  print(links[0])
```

```python
# Use of loop
article_title=[]
links=[]

for t in soup.select('h2.c-entry-box--compact__title a'):
  article_title.append(t.get_text())
  links.append(t.attrs.get('href'))

  print (article_title)
  print (links)
```

```python
# End Link
df = pd.DataFrame(
    {'article title': article_title,
     'links': links}
    )
print (df.head())

df.to_csv('vox.csv', index=False)
```
